```python
plantilla_datos = plantilla.merge(datos_personales, how = 'inner', left_on = 'Identificación del cliente (C.C / DNI / NIT / RUC)', right_on = 'Nro_Identificacion')
```

```python
for i in plantilla.index:
    #categorias del tipo de identificación
    if plantilla_datos.loc[i,'Tipo_Identificacion'] == 'CEDULA DE CIUDADANIA':
        plantilla.loc[i,'Tipo de identificación'] = 1
    else:
        plantilla.loc[i,'Tipo de identificación'] = plantilla_datos.loc[i,'Tipo_Identificacion']
    # categorias del genero
    if plantilla_datos.loc[i, 'Genero'] == 'Masculino':
        plantilla.loc[i, 'Género del cliente'] = 1
    elif plantilla_datos.loc[i, 'Genero'] == 'Femenino':
        plantilla.loc[i, 'Género del cliente'] = 2
    else:
        plantilla.loc[i, 'Género del cliente'] = 3
    #Categorias de la edad 
    if plantilla_datos.loc[i, 'Edad_y'] < 20:
        plantilla.loc[i,'Edad'] = 1
    elif 20 <= plantilla_datos.loc[i,'Edad_y'] <= 29:
        plantilla.loc[i,'Edad'] = 2
    elif 30 <= plantilla_datos.loc[i,'Edad_y'] <= 39:
        plantilla.loc[i,'Edad'] = 3
    elif 40 <= plantilla_datos.loc[i,'Edad_y'] <= 49:
        plantilla.loc[i,'Edad'] = 4
    elif 50 <= plantilla_datos.loc[i,'Edad_y'] <= 59:
        plantilla.loc[i,'Edad'] = 5
    elif plantilla_datos.loc[i, 'Edad_y'] >= 60:
        plantilla.loc[i,'Edad'] = 6
    else: 
        plantilla.loc[i,'Edad'] = 99
    #telefono del cliente
    plantilla.loc[i, 'Teléfono del cliente'] = plantilla.loc[i, 'Celular del cliente']
    #ciudad y departamento de residencia
    plantilla.loc[i,'Departamento/Región'] = plantilla_datos.loc[i,'Codigo_CIIUU_Ciudad_Residencia'][:2]
    plantilla.loc[i,'Ciudad/Provincia/Municipio'] = plantilla_datos.loc[i,'Codigo_CIIUU_Ciudad_Residencia'][3:]
    #CIIU
    plantilla.loc[i,'CIUU'] = plantilla_datos.loc[i,'Codigo_CIIU_Asociado']
    
```
