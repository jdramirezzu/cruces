##### Parte 2: Revisión de cancelación anticipada de CDATs y cofiahorros

```python
# Filtros del reporte
productos_revision = movimientos['Producto'].isin(['COAH - COFIAHORRO', 'CAHR - COFIAHORRITO','COSE - COFIAHORRO SEMILLA', 
                                      'CONT - DEPOSITO CONTRACTUAL', 'CISE - COFIAHORRITO SEMILLA', 
                                      'CDAT EMITIDOS MAYOR A 90 DIAS','CDSE - CDAT MAYOR A 90 DIAS SEMILLA' ])
idx = [i for i, val in enumerate(productos_revision) if not val]
productos_rev = movimientos.drop(idx)
```

```python
mergefechas = productos_rev.merge(ahorros[['depositos_numero_cuenta','depositos_fecha_inicio', 'depositos_fecha_vencimiento']], how = 'left', left_on = 'Nro_Cuenta', right_on = 'depositos_numero_cuenta')
mergefechas.drop('depositos_numero_cuenta', axis = 1)
```

```python
#Convertir columna de fechas en formato correcto
mergefechas['Fecha_Movimiento'] = pd.to_datetime(mergefechas['Fecha_Movimiento'])
mergefechas['depositos_fecha_inicio'] = pd.to_datetime(mergefechas['depositos_fecha_inicio'])
mergefechas['depositos_fecha_vencimiento'] = pd.to_datetime(mergefechas['depositos_fecha_vencimiento'])
```

```python
#Filtros necesarios para definir que un movimiento fue normal
iniciocdat = mergefechas['Fecha_Movimiento'] == mergefechas['depositos_fecha_inicio']
fincdat1 = mergefechas['Fecha_Movimiento'] == mergefechas['depositos_fecha_vencimiento']
fincdat2 = (mergefechas['Fecha_Movimiento'] - mergefechas['depositos_fecha_vencimiento']) > np.zeros((mergefechas.shape[0]))
fincdat3 = (mergefechas['depositos_fecha_vencimiento'] - mergefechas['Fecha_Movimiento']) <= np.ones((mergefechas.shape[0]))*3
saldofinalcdat = mergefechas['Saldo_Final'] > np.zeros((mergefechas.shape[0]))
saldofinalvsinicial = mergefechas['Saldo_Final'] >= mergefechas['Saldo_Inicial']
movmasinicial = (mergefechas['Monto_Movimiento'] + mergefechas['Saldo_Inicial']) == mergefechas['Saldo_Final']
```

```python
#Condiciones para definir si un movimiento fue normal
condiciones = iniciocdat | fincdat1 | (fincdat2 & fincdat3) | (saldofinalcdat & saldofinalvsinicial) | movmasinicial
# Encontrar indice de movimientos normales
indices = [i for i, val in enumerate(condiciones) if not val]
#Seleccionar registros anormales
mergefechas = mergefechas.loc[indices, :]
```

```python
mergefechas
```

```python
for i in mergefechas.index:
    Cdat = mergefechas.loc[i, 'Nro_Cuenta']
    Client = mergefechas.loc[i, 'Cliente']
    Cc = mergefechas.loc[i, 'Cedula']
    Fecha = mergefechas.loc[i, 'Fecha_Movimiento']
    Vencimiento = mergefechas.loc[i, 'depositos_fecha_vencimiento']
    mergefechas.loc[i, 'Redmine'] = ('Se solicita verificar si existe autorización para cancelación de CDAT {0} de {1} - {2} cancelado el día {3}, cuya fecha de vencimiento era {4}'.format(Cdat, Client, Cc, Fecha, Vencimiento))
```

```python
mergefechas = mergefechas.drop(['depositos_numero_cuenta', 'depositos_fecha_inicio', 'depositos_fecha_vencimiento'], axis=1)
Cruce1 = pd.concat([mov, mergefechas])
```

#### Cruce 2: Revisión de consignación a aportes, cuota de admisión e iva.

```python
#Leer Reporte de Datos Personales
datospersonales = pd.read_excel('personal prueba cathe.xlsx', usecols = ['Agencia', 'Tipo_Identificacion', 'Nro_Identificacion',
                                                                   'Nombre_Completo', 'Genero', 'Fecha_Nacimiento', 'Edad',
                                                                   'Nacionalidad', 'Tipo_Persona', 'Vinculo', 'Estrato',
                                                                    'Direccion_Residencia', 'Celular', 'Email', 'Fecha_Ingreso',
                                                                    'Fecha_Reingreso1', 'Usuario_Ultima_Actualizacion'])
```

```python
#Filtro reporte basico de ahorros 
ahorros_activos = ahorros.loc[:, 'depositos_estado'] == 'Activo'
aportesord = ahorros['depositos_nombre_producto'].isin(['APORTES ORDINARIOS'])
aportesord_activos = ahorros_activos & aportesord
idxaportes = [i for i, val in enumerate(aportesord_activos) if val]
```

```python
aportesord_activo = ahorros.loc[idxaportes, :]
```

```python
#Leer auxiliar que contiene cuota de admision e iva
aux = pd.read_excel('comprobante soporte canales.xlsx')
```

```python
# Definir rango de fecha de revision
a = movimientos['Fecha_Movimiento'].unique()[0]
b = pd.datetime.today()
fecha = pd.date_range(start=a,end=b)
```

```python
#Filtrar datos personales (reporte total) pro fecha de ingreso (fecha a revisar)
indice = []
for i in datospersonales.index:
    if datospersonales.loc[i, 'Fecha_Ingreso'] in fecha:
        indice.append(i)
```

```python
# Filtrado anterior. Modificacion de tipo de datos de numero de identificacion ya que esta se lee como un objeto en el csv
dp = datospersonales.loc[indice, :]
dp['Nro_Identificacion'] = dp['Nro_Identificacion'].astype(float)
```

```python
#Filtrado por vinculo
idxsocio = []
for i in dp.index:
    if ('Socio' in dp.loc[i, 'Vinculo']) & ('Cliente' in dp.loc[i, 'Vinculo']):
        idxsocio.append(i)
```

```python
dp = dp.loc[idxsocio, :]
```

```python
# Extraer aux de cuota de admision e iva 
cuotaadmon = (aux['ACCTVALUE'] == 42300501)
iva = aux['ACCTVALUE'] == 2440100101
indicescuota = [i for i, val in enumerate(cuotaadmon) if val]
indicesiva = [i for i, val in enumerate(iva) if val]
auxcuota = aux.loc[indicescuota, :]
auxiva = aux.loc[indicesiva, :]
```

```python
#Unir auxiliar de cuota de admision con datos personales usando numero de cedula
asociaciones = dp.merge(auxcuota[['IDENTIFICACION','CREDITO', 'ACCTVALUE', 'NUM_DOC']], how = 'left', left_on = 'Nro_Identificacion', right_on = 'IDENTIFICACION')
asociaciones = asociaciones.drop('IDENTIFICACION', axis=1)
asociaciones
```

```python
#Unir archivo generado en paso anterior con iva usando numero de cedula y numero de documento , esto porque existen
#muchos registros de iva para una misma cedula
asociaciones = asociaciones.merge(auxiva[['IDENTIFICACION','CREDITO', 'ACCTVALUE', 'NUM_DOC']], how = 'left', left_on = ['Nro_Identificacion', 'NUM_DOC'], right_on = ['IDENTIFICACION', 'NUM_DOC'])
```

```python
#Unir archivo generado en paso anterior con valor en aportes ordinarios usando numero de cedula 
asociaciones = asociaciones.merge(aportesord_activo[['identificacion','depositos_saldo_actual']], how = 'left', left_on = ['Nro_Identificacion'], right_on = ['identificacion'])
```

```python
asociaciones
```

```python
#Verificaciones: cuota de admision mayor a $7315. Iva equivalente al 19% de la cuota cobrada. Si no cumple con estas
#condiciones se genera una columna con alerta
aportes_val = asociaciones['depositos_saldo_actual'] >= 49687
cuota_val = asociaciones['CREDITO_x'] >= 7315
iva_val1 = asociaciones['CREDITO_y'] >= ((asociaciones['CREDITO_x']*0.19)-20)
iva_val2 = asociaciones['CREDITO_y'] < ((asociaciones['CREDITO_x']*0.19)+20)
verif = aportes_val & cuota_val & iva_val1 & iva_val2
idxverif = [i for i, val in enumerate(verif) if not val]
```

```python
Cruce2 = asociaciones.loc[idxverif, :]
```

```python
Cruce2
```

```python
#Mensaje de error 
for i in Cruce2.index:
    Cc = Cruce2.loc[i, 'Nro_Identificacion']
    Nombre = Cruce2.loc[i, 'Nombre_Completo']
    Ingreso = Cruce2.loc[i, 'Fecha_Ingreso']
    Saldoaportes = Cruce2.loc[i, 'depositos_saldo_actual']
    Cuota_adm = Cruce2.loc[i, 'CREDITO_x']
    Iva = Cruce2.loc[i, 'CREDITO_y']
    if not Cruce2.loc[i, 'CREDITO_x'] >= 7315:
        Cruce2.loc[i, 'Redmine'] = ('La cuota de admisión de {0} - {1}, ingresado/a a datos personales el {2}, fue por ${3} inferior al valor establecido'.format(Cc, Nombre, Ingreso, Cuota_adm))
    elif not Cruce2.loc[i, 'depositos_saldo_actual'] >= 49687:
        Cruce2.loc[i, 'Redmine'] = ('Se solicita consignar aportes ordinarios de asociación de {0} - {1}, ingresado/a a datos personales el {2}.'.format(Cc, Nombre, Ingreso))
    elif not (Cruce2.loc[i, 'CREDITO_y'] >= ((Cruce2.loc[i, 'CREDITO_x']*0.19)-20)) and (Cruce2.loc[i, 'CREDITO_y'] < ((Cruce2.loc[i, 'CREDITO_x']*0.19)+20)):
        Cruce2.loc[i, 'Redmine'] = ('El valor de iva recibido por esta asociación es inferior al 19% de la cuota de admisión de {0} - {1} del {2}, fue por ${3} inferior al 19% de la cuota de admisión por {4}'.format(Cc, Nombre, Ingreso, Iva, Cuota_adm))
```

```python
#Exportar resultados en un único archivo de excel. Una hoja por cada Cruce.
today = date.today()
filename = 'CRUCES {}.xlsx'.format(today)
writer = pd.ExcelWriter(filename, engine='xlsxwriter')
Cruce1.to_excel(writer, sheet_name='1', index=False)
Cruce2.to_excel(writer, sheet_name='2', index=False)
workbook  = writer.book
worksheet1 = writer.sheets['1']
worksheet2 = writer.sheets['2']
```
